```python
import torch.nn as nn

class LinearRegressionModel(nn.Module):
    def __init__(self, input_dim, output_dim):
        super(LinearRegressionModel, self).__init__()
        self.linear = nn.Linear(input_dim, output_dim, bias=True)

    def forward(self, x):
        return self.linear(x)
```

```python
import torch
import pickle
DRIVE_PATH = "/content/drive/MyDrive/Hackathon: Contrastive SAE Steering"

with open(f'{DRIVE_PATH}/training_tensors.pkl', 'rb') as f:
    X_tensor_loaded, y_tensor_loaded = pickle.load(f)
```

```python
input_dim = 25
output_dim = 10
reggression_model = LinearRegressionModel(input_dim, output_dim)
X = X_tensor_loaded.clone()
y = y_tensor_loaded.clone()

# Define the loss function and optimizer
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(reggression_model.parameters(), lr=0.01)

# Training loop
num_epochs = 1000

for epoch in range(num_epochs):
    reggression_model.train()

    # Forward pass
    outputs = reggression_model(X)
    loss = criterion(outputs, y)

    # Backward pass and optimization
    optimizer.zero_grad()  # Zero the gradients
    loss.backward()  # Compute gradients
    optimizer.step()  # Update weights

    if (epoch + 1) % 100 == 0:
      print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}')
```

```python
# Get the learned weights
weights = reggression_model.linear.weight.data
intercept = reggression_model.linear.bias
print(weights)
print(intercept)
```

```python
with open(f'{DRIVE_PATH}/weight_tensors.pkl', 'wb') as f:
    pickle.dump((weights, intercept), f)
```
